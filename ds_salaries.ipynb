#Visualizing the Data Science Jobs dataset using Matplotlib

```python
# Count the occurrences of each job title
job_title_counts = df['job_title'].value_counts()

# Select the top most job titles (you can change the number according to your preference)
top_job_titles = job_title_counts.head(10)

# Sort the job titles by count in descending order
top_job_titles_sorted = top_job_titles.sort_values(ascending=True)

# Plotting the horizontal bar plot
plt.figure(figsize=(10, 6))
top_job_titles_sorted.plot(kind='barh', color='m')
plt.title('Top Most Job Titles in Data Science', fontweight="bold",color='r')
plt.xlabel('Count', fontweight="bold",color='black')
plt.ylabel('Job Title', fontweight="bold",color='black')
plt.tight_layout()
```

Insights:

---


The horizontal bar graph represents the top most job titles in data science
based on their respective counts.
Data Engineer, The most frequent job title in data science, indicating a high demand for professionals

Data Scientist and Data Analyst, another popular job titles, suggesting a need for professionals skilled in analyzing and interpreting data to support decision-making processes.

```python
# Calculate the frequency of each job title
job_title_counts = df['job_title'].value_counts().head(10)

# Plotting the top 10 job titles
plt.figure(figsize=(8,6))
job_title_counts.plot(kind='bar')
plt.title('Top 10 Job Titles in 2023',fontweight="bold",color='r')
plt.xlabel('Job Title',fontweight="bold",color="black")
plt.ylabel('Frequency',fontweight="bold",color="black")
plt.xticks(rotation=90)
plt.tight_layout()
```

Insights:


---
The bar graph represents the top 10 job titles in data science based on the year 2023.

We can visualize that their was a dominance of Data Engineer Jobs followed by the Data Scientists and Data Analysts.



```python
# Grouping data by work year and job title, then counting occurrences
job_titles_by_year = df.groupby(['work_year', 'job_title']).size().unstack(fill_value=0)
# Summing up occurrences of each job title over the years
job_title_totals = job_titles_by_year.sum(axis=0)
# Selecting top 20 job titles
top_20_job_titles = job_title_totals.sort_values(ascending=False).head(10).index
# Filtering the data to include only the top 20 job titles
job_titles_by_year_top_20 = job_titles_by_year[top_20_job_titles]
plt.figure(figsize=(10, 5))
# Plotting each of the top 20 job titles
for job_title in job_titles_by_year_top_20.columns:
    plt.bar(job_titles_by_year_top_20.index.astype(int), job_titles_by_year_top_20[job_title], label=job_title)
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Work Year', fontweight="bold", color="black")
plt.ylabel('Number of Occurrences', fontweight="bold", color="black")
plt.title('Leading Job Titles Over the Years', fontweight="bold", color='r')
plt.xticks(job_titles_by_year_top_20.index.astype(int), ['2020', '2021', '2022', '2023'])
plt.tight_layout()
```

Insights:

---
 To analyze the top job titles by their frequency over the work years. This will give us a focused look at the most common job titles in the dataset and how their occurrence has changed from 2020 to 2023.

 Data Analyst is the most in-demand job-title follwed by the Data Scientist and the Data Engineer.
 The demand for these jobs has been increasing over the past years.



```python
# Selecting top 10 highest paid roles in data science
top10_roles = df.groupby('job_title', as_index=False)['salary'].max().sort_values(by='salary', ascending=False).head(10)

# Create a horizontal bar plot
plt.figure(figsize=(10, 5))
bars = plt.barh(top10_roles['job_title'], top10_roles['salary'], color='blue')

# Set plot title and labels
plt.title('Highest Paid Roles in Data Science', fontsize=18, fontweight='bold')
plt.xlabel('Salary in USD', fontsize=14, fontweight='bold')
plt.ylabel('', fontsize=14, fontweight='bold')  # No label for y-axis
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add data labels
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"${bar.get_width():,.0f}", va='center')

plt.tight_layout()
```

Insights:

---
To have a look over salaries in data science, the plot is here..
We can visualize that role of data scientist has the maximum salary followed by others

```python
# Counting the occurrences of each experience level
experience_counts = df['experience_level'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(9, 6))
ex=[0.1,0,0,0]
plt.pie(experience_counts, labels=experience_counts.index, autopct='%1.1f%%', startangle=140,explode=ex)
plt.title('Experience Level Distribution', fontsize=18, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle


# Adding a legend
plt.legend(title='Experience Level', loc='upper right')

plt.tight_layout()
plt.show()
```

Pie chart depicts the distribution of experience levels in the given dataset . We can analyze the senior level are the maximum ones that are eomployed in the data science.

```python
# Filter the data for salaries in USD
df_usd = df[df['salary_currency'] == 'USD']

# Plotting the box plot
plt.figure(figsize=(10, 6))
plt.boxplot([df_usd[df_usd['experience_level'] == exp]['salary'] for exp in df_usd['experience_level'].unique()],
            labels=df_usd['experience_level'].unique(), patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Salary Distribution by Experience Level', color='r', fontweight='bold')
plt.xlabel('Experience Level', fontweight='bold')
plt.ylabel('Salary ($)', fontweight='bold')
plt.xticks(rotation=360, ha='right')
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

Insights:

---



By looking at the measures we can get a good idea how salaries vary by experience level. So the expert level(ex) experience is having the highest salary followed by the senior level.

```python
# Grouping data by employment type and calculating median salary
median_salary_by_employment_type = df.groupby('employment_type')['salary'].mean().sort_values()

# Plotting the bar plot
plt.figure(figsize=(8, 5))
bars = plt.bar(median_salary_by_employment_type.index, median_salary_by_employment_type.values, color='purple',width=0.5)
plt.title('Average Salary by Employment Type', fontsize=16, fontweight='bold', color='red')
plt.xlabel('Employment Type', fontsize=14, fontweight='bold', color='black')
plt.ylabel('Average Salary', fontsize=14, fontweight='bold', color='black')
plt.xticks(rotation=360)  # Rotate x-axis labels for better readability

# Adding annotations to the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, '{:.2f}'.format(height),
             ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

Insights:

---


So the Freelancer is earning the highest followed by the fulltime

```python
# Replace values in the DataFrame
df['company_size'].replace({'L': 'Large', 'M': 'Medium', 'S': 'Small'}, inplace=True)

# Calculate value counts for company size
company_size_counts = df['company_size'].value_counts()

# Plotting the doughnut chart
plt.figure(figsize=(5, 6))

# Outer pie chart (doughnut)
outer_colors = ['r', 'g', 'b']
plt.pie(company_size_counts, labels=company_size_counts.index, autopct='%1.1f%%', startangle=140,
        colors=outer_colors, pctdistance=0.85, wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner circle to create the doughnut
inner_circle = plt.Circle((0, 0), 0.5, color='white')
plt.gca().add_artist(inner_circle)

# Adding legend
plt.legend(title='Company Size', loc='upper right')

plt.title('Distribution of Company Size', fontsize=18, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
```

Insights:

---


The Meduim size companies are the highly growable providing the data science jobs

```python
# Define salary ranges and labels
salary_bins = [0, 50000, 100000, 150000, 200000, 250000, 300000, float('inf')]
salary_labels = ['< 50k', '$50k - $100k', '$100k - $150k', '$150k - $200k', '$200k - $250k', '$250k - $300k', '> 300k']

# Bin salaries into salary ranges
df['salary_range'] = pd.cut(df['salary'], bins=salary_bins, labels=salary_labels)

# Group data by company size and salary range, then calculate count
salary_distribution = df.groupby(['company_size', 'salary_range']).size().unstack(fill_value=0)

# Plot grouped bar chart for each company size category
salary_distribution.plot(kind='bar', figsize=(8, 6))
plt.title('Distribution of Salaries by Company Size',fontweight="bold",color='r')
plt.xlabel('Company Size',fontweight="bold")
plt.ylabel('Count',fontweight="bold")
plt.xticks(rotation=360)
plt.legend(title='Salary Range', bbox_to_anchor=(1.05, 1), loc='upper left')

```

Insights:

---
We can analyze the meduim size companies offers the higher salaries


```python
# Grouping data by experience level and company size
explevel_size = df.groupby(['experience_level', 'company_size']).size().unstack()

# Set the bar width
bar_width = 0.2

# Set the position of each group on the X-axis
r1 = range(len(explevel_size.columns))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Create the figure and axes
plt.figure(figsize=(10, 5))

# Plotting the bars for each experience level
plt.bar(r1, explevel_size.loc['EN'], color='blue', width=bar_width, edgecolor='grey', label='Entry-Level')
plt.bar(r2, explevel_size.loc['EX'], color='orange', width=bar_width, edgecolor='grey', label='Expert-Level')
plt.bar(r3, explevel_size.loc['MI'], color='green', width=bar_width, edgecolor='grey', label='Mid-level')
plt.bar(r4, explevel_size.loc['SE'], color='m', width=bar_width, edgecolor='grey', label='Senior-Level')

# Add xticks and labels
plt.xlabel('Company Size', fontweight='bold')
plt.ylabel('Records', fontweight='bold')
plt.title('Experience level and company size', fontsize=18, fontweight='bold')
plt.xticks([r + bar_width for r in range(len(explevel_size.columns))], explevel_size.columns)

# Add legend
plt.legend()

plt.tight_layout()
plt.show()
```

Insights:

---
We can analyze here the senior and mid level experts are more predominant in meduin company sizes.


```python
# Counting the occurrences of each remote ratio
remote_ratio_counts = df['remote_ratio'].value_counts()


# Plotting the doughnut chart
plt.figure(figsize=(5, 7))

# Outer pie chart (doughnut)
outer_colors = ['blue', 'green', 'coral']
plt.pie(remote_ratio_counts, labels=remote_ratio_counts.index, autopct='%1.1f%%', startangle=140,
        colors=outer_colors, pctdistance=0.85, wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner circle to create the doughnut
inner_circle = plt.Circle((0, 0), 0.5, color='white')
plt.gca().add_artist(inner_circle)

# Adding legend
plt.legend(title='Remote Ratio', loc='upper right')

plt.title('Distribution of Remote Ratio', fontsize=18, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
```

Insights:

---
This informs us the various percentages of various of the full remote work,no remote work and partial remote work


```python
# Grouping data by work year and remote ratio, then counting company locations
df_remote_ratio_by_year = df.groupby(['work_year', 'remote_ratio']).size().unstack(fill_value=0)
plt.figure(figsize=(10,5))

# Plotting each work year
for i, (work_year, row) in enumerate(df_remote_ratio_by_year.iterrows()):
    for j, remote_ratio in enumerate(df_remote_ratio_by_year.columns):
        plt.bar(i + j * 0.8 / len(df_remote_ratio_by_year.columns), row[remote_ratio],
                width=0.8 / len(df_remote_ratio_by_year.columns), color=plt.cm.tab10(j), alpha=0.7)

# Adding legend
plt.legend(df_remote_ratio_by_year.columns, title='Remote Ratio', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding labels and title
plt.xlabel('Work Years',fontweight="bold", color='black')
plt.ylabel('Count of Company Locations',fontweight="bold", color='black')
plt.title('Working Model Followed per Work Year',fontweight="bold",color='r')

# Setting x-axis ticks to work years and rotating for better readability
plt.xticks(range(len(df_remote_ratio_by_year)), df_remote_ratio_by_year.index, rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

Insights:

---
We can see their was a hype of remote work in the year 2021 and 2022 , then in the year 2023 thier is a hype in no remote work followed by fully remote work


```python
# This is a small subset and should be expanded based on the actual country codes present in the dataset
country_code_mapping = {
    'US': 'United States',
    'IN': 'India',
    'GB': 'United Kingdom',
    'DE': 'Germany',
    'CA': 'Canada',
    'FR': 'France',
    'ES': 'Spain',
    'JP': 'Japan',
    'IT': 'Italy',
    'AU': 'Australia',
    'BR': 'Brazil',
    'GR':'Greece',
    'PT': 'Portugal',
}

# Replace the country codes in the dataset with full country names using the mapping
df['employee_residence_full'] = df['employee_residence'].replace(country_code_mapping)

# Group the data by the full employee residence name and calculate the average remote ratio for each
grouped_residence_data = df.groupby('employee_residence_full')['remote_ratio'].mean().reset_index()
# Select the top 10 residences with the highest number of job opportunities
top_residences = df['employee_residence_full'].value_counts().head(10).index

top_residences_data = grouped_residence_data[grouped_residence_data['employee_residence_full'].isin(top_residences)]
# Sort the top residences by the average remote ratio
top_residences_data_sorted = top_residences_data.sort_values(by='remote_ratio', ascending=False)

# Create a bar plot for the top 10 residences and their average remote ratio
plt.figure(figsize=(10, 6))
plt.scatter(top_residences_data_sorted['employee_residence_full'], top_residences_data_sorted['remote_ratio'], color='red')
plt.title('Average Remote Ratio for Top 10 Employee Residences',color='Blue',fontweight="bold")
plt.xlabel('Employee Residence',fontweight='bold')
plt.ylabel('Average Remote Ratio (%)',fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
```

Insights:


---
here we can visualize the contries which prefer most remote options are Brazil and Greece

```python
# This is a small subset and should be expanded based on the actual country codes present in the dataset
country_code_mapping = {
    'US': 'USA',
    'GB': 'UK',
    'CA': 'Canada',
    'DE': 'Germany',
    'IN': 'India',
    'ES':'Spain',
    'FR':'France',
    'AU':'Australia',
    'BR':'Brazil' ,
    'GR':'Greece',
}

# Replace the country codes in the dataset with full country names using the mapping
df['company_location_full'] = df['company_location'].replace(country_code_mapping)

# Grouping data by the full company location name and counting the number of job opportunities
company_locations_full_counts = df['company_location_full'].value_counts()

# Selecting the top 10 company locations with full names
top_10_company_locations_full = company_locations_full_counts.head(10)

# Plotting the top 10 company locations with the most job opportunities using full country names
plt.figure(figsize=(10,5))
top_10_company_locations_full.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Top 10 Company Locations with Most Job Opportunities', color='r',fontweight='bold')
plt.xlabel('Company Location',fontweight='bold')
plt.ylabel('Number of Job Opportunities',fontweight='bold')
plt.xticks()
plt.tight_layout()
```
